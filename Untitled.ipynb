```python
#Import that matplotlib dependencies.

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()

# Add label, change the line color to gree, the marker to a diamond, the linewidth to 2, and the chart label to Boston

ax.plot(x_axis, y_axis, color = "green", marker = "D", linewidth=2, label='Boston' )

# Set x label to "Date"
ax.set_xlabel("Date")

# Set y label to "Fare($)"
ax.set_ylabel("Fare($)")

# Set title to "PyBer Fare by Month"
ax.set_title("PyBer Fare by Month")

# Set y limit to 0 thru 45
ax.set_ylim(0, 45)

# Add grid lines
ax.grid()
```

```python
# Create the plot with ax.plt()

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot

plt.plot(x_axis, y_axis)

plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axis

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45

plt.ylim(0, 45)

# Create a title

plt.title("PyBer Fare by Month")

# Add a legend

plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.barh(y_axis,x_axis, color="green", label='Boston')

# Create labels for the x and y axis

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title

plt.title("PyBer Fare by Month")

# Add a legend

plt.legend()
```

```python
# Create the plot

plt.barh(x_axis,y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color = "cyan", label = "Boston")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fair($)")
ax.legend()
```

```python
plt.scatter(y_axis,x_axis, color = "red", label = "Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "skyblue", linewidth =2, label = "Boston", alpha = 0.8, edgecolor = "black", s=[i * 5 for i in y_axis])
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis, colors = colors, explode = explode_values, autopct='%.1f%%', shadow='True', startangle = 85)
ax.invert_xaxis()
plt.show()
```
